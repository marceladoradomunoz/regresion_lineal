# **PREGUNTA Nº 3**: **CONSTRUYE UN MODELO DE REGRESIÓN LINEAL QUE RELACIONE EL VOLUMEN Y EL PESO DE LOS PECES. REALIZA UNA BREVE EVALUCIÓN DEL MODELO Y GRAFÍCALO**.

```python
import statsmodels.api as sm


# Definir las variables predictoras (X) y la variable de respuesta (y)
volumen = df_fish_species['Volume']
peso = df_fish_species['Weight']

# Agregar una constante al predictor (intercepto)
X = sm.add_constant(volumen)

# Crear el modelo de regresión lineal
modelo = sm.OLS(peso, X).fit()

# Obtener los resultados del modelo
resultados = modelo.summary()

# Imprimir los resultados
print(resultados)

# Graficar la regresión lineal
plt.figure(figsize=(10, 6))
plt.scatter(volumen, peso, label='Datos', color='pink')
plt.plot(volumen, modelo.predict(X), color='red', label='Regresión Lineal')
plt.title("Regresión lineal entre el volumen y el peso de los peces")
plt.xlabel("Volumen de los peces")
plt.ylabel("Peso de los peces")
plt.legend()
plt.grid(True)
plt.show()
```

# **EXPLICACIÓN DEL MODELO**: El modelo de regresión lineal ordinaria creado tiene por objeto es intentar predecir el peso de los peces (variable dependiente) en función de su volumen (variable independiente).

# **REVISIÓN DE LOS RESULTADOS**:


1.   **R-cuadrado:** Éste se ajusta a los datos, siendo de 0.778, significando que aproximadamente un 77.8% de la variabilidad en el peso de los peces se explica por el volumen.
2.   **Coeficiente de la regresión:** la ecuación entre el peso más el volumen multiplicado por el volumen otorga un resultado de un 58.7401, lo que significa que cuando el volumen tenga un valor de 0, se esperaría que el peso del pez se aproximadamente de un 58.74.
3.   **Coeficinete de volumen (variable "volume"):** Este arroja un valor de 1.7396, significando que, en promedio, por cada unidad adicional de volumen, se esperaría un aumento de 1.7396 unidades en el peso.
4.   **P-values:** el valor de p otorgado paqra cada coeficiente de prueba, confirma la hipótesis nula de que el coeficiente es igual a cero (en otras palabras, no tiene un efecto en la predicción). Ambos p-values son muy bajos, específicamente, cercanos a 0, lo que indicaría que tanto el coeficiente de constante como el coeficiete de volumen son estadísticamente significativos.
5.   **F-Statistic:** Siendo la medida global de la significancia de relación entre las variables independientes y dependientes, y arrojanbdo un valor de 549.9, sumado al valor de probabilidad (dado por Prob), estaría indicando que existe un 549l.9 de aproximarse 3.65 (notación científica). Aquello significa que el modelo creado, en general, es estadísticamente significativo.
6.   **Error estándar:** Considerando que a más pequeño el error estándar, más precisa serán las estimaciones de los coeficientes.

# **CONCLUSIÓN**: En base a lo anterior, podemos visualizar la existencia de linealidad entre las variabables "Volume" y "Weght" de los peces. En otras palabras, en el modelo precedente, podemos visualizar que existe linealidad entre las variables previamente señaladas. El R-cuadrado nos indica que el modelo se ajusta bien a los datos en relación con el Estadístico F y el valor P. Por ende, se estaría en presencia de un modelo signitictivo.



# **PREGUNTA Nº 4**: **QUÉ SUCEDE SI APLICAS EL MODELO ANTERIOR POR SEPARADO A CADA ESPECIE DE PECES? EXPLICA**.


```python
#CREACIÓN DEL CÓDIGO QUE PERMITE REALIZAR UN ANÁLISIS DE REGRESIÓN LINEAL MÚLTIPLE CON USO DE
#VARIABLES PREDICTORAS DONDE SE INCLUYEN DATOS NUMÉRICOS Y CATEGÓRICOS.

# Paso 1: Definición de la variable predictora (X) y de la variable de respuesta (y).

X = df_fish_species[['Volume', 'Species']]  #Volumen del pez y especies como variables independientes
y = df_fish_species['Weight']  #Peso como variable dependiente

# Codificació de la variable categórica 'Species' usando get dummies.
X_encoded = pd.get_dummies(X, columns=['Species'], prefix=['Species'])

# Paso 2: Agregagación de una constante (intercepto) a las variables predictoras.
X_encoded = sm.add_constant(X_encoded)

# Paso 3: Creación del modelo de regresión lineal.
model = sm.OLS(y, X_encoded.astype(float))
model = model.fit()

# Paso 4: Obteneción de los resultados del modelo creado.
results = model.summary()
predictions = model.predict(X_encoded)

#Calculo del Error Cuadrático Medio (MSE).
mse = ((y - predictions) ** 2).mean()

#Calculo del Error Cuadrático Medio Explicado (EMSE).
emse = mse / ((y - y.mean()) ** 2).mean()

#Obtención del valor R² del modelo creado.
r2 = model.rsquared

print(results)
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Error Mean Squared Explicado (EMSE): {emse:.2f}")
print(f"Coeficiente de Determinación (R²): {r2:.2f}")
```

# **EXPLICACIÓN DEL MODELO CREADO**: Se crea un modelo de regresión linean múltiple cuyo objeto es el intento de predicción del peso de los peces (variable dependiente), en el cual se utilizan diversas variables predictoras (volumen y especie del pez codificadas como variables dummy).

# **REVISIÓN DE LOS RESULTADOS**:


1.   **R-cuadrado**: Arroja un valor de 0.899, lo que implicaqría que aproximadamente el 89.9% de la variabiidad delpeso de los peces puede ser explicado por las variables que se incluyen en el modelo de regresión lineal múltiple.
2.   **Coeficiente de regresión:** En el caso en concreto, cada coeficiente representa el cambio esperado de la variable dependiente (peso de los peces) por medio de un cambio unitario en la variable independiente. El coeficiente para "Volume" es de 1.8306, lo que significaría que en promedio se esperaría un aumento de 1.8306 unidades en el peso de los peces por cada unidad adicional de volumen. Los coeficientes arrojadas para las variables dummy ("Species") representarían el cambio en el peso de los peces según su especie.
3.   **P-values:** Los valores p de cada coeficiente nos entrega la medida de significancia estadística para cada coeficiente. Se ha de considerar que un valor p bajo (menor a 0.5) indicaría que el coeficiente es significativamente diferente a 0.
      En el caso de la variable "Volume", el valor p es muy bajo (menor a 0.001), e indicaría que el coeficiente es significativamente diferente de 0 y que el volumn es una variable predictiva importante para el peso de los peces.
4.   **F-statistic y probabilidad**: El F-statistic es de 191.5 con un valor prob de casi 0, y ello indicaría que el modelo creado, en general, es, estadísticamente significativo.
5.   **Error estándar:** Al parecer, los resiguods no siguen una distribución normal.
6.   Notas: nos otorgan información adicional en cuanto a las especificaciones del modelo creado y probables inconvenientes que pueden presentarse con un modelo de regresión lineal múltiple.
7.   Coeficiente de Determinación: Nos indica la proporción de la varianza en la variable dependiente, arrojando un valor de 0.90, lo que indica que el modelo se ajusta a los datos.

# **CONCLUSIÓN**: Analizando someramente los datos que arroja el modelo, estos demuestran que el mismo tiene un buen ajuste a los datos ingresados, sumado a que las variables predictoras que se introdujeron resultan significativas para predecir el peso de los peces. Sin embargo, si bien esta parte no tiene la capacidad para comprender a cabalidad la información que arrojan las notas, sí se comprende que se ha de tener en consideración que pueden existir problemas con la multicolinealidad y el diseño de la matriz.
     En general, podemos visualizar claramente que el modelo logra separar los peces según su especie y muestra los valores para cada una de ellas (las 7). # **PREGUNTA Nº 5**: **CONSTRUYE UN MODELO DE REGRESIÓN QUE RELACIONE DOS O MÁS VARIABLES CON EL PESO DE LOS PECES. REALIZA UNA BREVE EVALUACIÓN DEL MODELO**.


```python
#REVISIÓN VARIALES "SPECIES" A OBJETO DE VISUALIZAR LAS DISTINTAS ESPECIES DE PECES EXISTENTES EN EL DF.

df_fish_species['Species'].unique()
```

```python
#PASOS A SEGUIR
#1.- Revisión de filas y eliminación de valores faltantes.
#2.- Filtrar los datos según especie de pez.
#3.- Creación de modelos de regresión lineal para cada género y ambos de manera conjunta.
#4.- Obtención de las predicciones para cada modelo.
#5.- Cálculo del error cuadrático medio (MSE) para cada modelo.
#6.- Cálculo del error cuadrático medio explicado (EMSE) para cada modelo
#7.- Impresión de los resultados.

#Eliminar filas con valores faltantes
data_fish_clean = df_fish_species.dropna(subset=['Weight', 'Width'])

# Filtrar los datos por especie (Roach, Whitefish,Smelt,Parkki)
pez_roach = data_fish_clean[data_fish_clean['Species'] == 'Roach']
pez_whitefish = data_fish_clean[data_fish_clean['Species'] == 'Whitefish']
pez_smelt = data_fish_clean[data_fish_clean['Species'] == 'Smelt']
pez_parkki = data_fish_clean[data_fish_clean['Species'] == 'Parkki']


# Crear modelos de regresión lineal para cada género y para ambos juntos
model_roach = sm.OLS(pez_roach['Weight'], sm.add_constant(pez_roach['Width'])).fit()
model_whitefish = sm.OLS(pez_whitefish['Weight'], sm.add_constant(pez_whitefish['Width'])).fit()
model_smelt = sm.OLS(pez_smelt['Weight'], sm.add_constant(pez_smelt['Width'])).fit()
model_parkki = sm.OLS(pez_parkki['Weight'], sm.add_constant(pez_parkki['Width'])).fit()
model_todos = sm.OLS(data_fish_clean['Weight'], sm.add_constant(data_fish_clean['Width'])).fit()


# Obtener las predicciones de cada modelo
predictions_roach = model_roach.predict()
predictions_whitefish = model_whitefish.predict()
predictions_smelt = model_smelt.predict()
predictions_parkki = model_parkki.predict()
predictions_todos = model_todos.predict()


# Calcular el Error Cuadrático Medio (MSE) para cada modelo
mse_roach = ((pez_roach['Weight'] - predictions_roach) ** 2).mean()
mse_whitefish = ((pez_whitefish['Weight'] - predictions_whitefish) ** 2).mean()
mse_smelt = ((pez_smelt['Weight'] - predictions_smelt) ** 2).mean()
mse_parkki = ((pez_parkki['Weight'] - predictions_parkki) ** 2).mean()
mse_todos = ((data_fish_clean['Weight'] - predictions_todos) ** 2).mean()


# Calcular el Error Cuadrático Medio Explicado (EMSE) para cada modelo
emse_roach = mse_roach / ((pez_roach['Weight'] - pez_roach['Weight'].mean()) ** 2).mean()
emse_whitefish = mse_whitefish / ((pez_whitefish['Weight'] - pez_whitefish['Weight'].mean()) ** 2).mean()
emse_smelt = mse_smelt / ((pez_smelt['Weight'] - pez_smelt['Weight'].mean()) ** 2).mean()
emse_parkki = mse_parkki / ((pez_parkki['Weight'] - pez_parkki['Weight'].mean()) ** 2).mean()
emse_todos = mse_todos / ((data_fish_clean['Weight'] - data_fish_clean['Weight'].mean()) ** 2).mean()


# Calcular el Coeficiente de Determinación (R²) para cada modelo
r2_roach = model_roach.rsquared
r2_whitefish = model_whitefish.rsquared
r2_smelt = model_smelt.rsquared
r2_parkki = model_parkki.rsquared
r2_todos = model_todos.rsquared

# Calcular el Error Absoluto Medio (MAE) para cada modelo
mae_roach = (abs(pez_roach['Weight'] - predictions_roach)).mean()
mae_whitefish = (abs(pez_whitefish['Weight'] - predictions_whitefish)).mean()
mae_smelt = (abs(pez_smelt['Weight'] - predictions_smelt)).mean()
mae_parkki = (abs(pez_parkki['Weight'] - predictions_parkki)).mean()
mae_todos = (abs(data_fish_clean['Weight'] - predictions_todos)).mean()

# Mostrar los resultados
print("Modelo para Peces Roach:")
print(f"R²: {r2_roach:.2f}, MSE: {mse_roach:.2f}, EMSE: {emse_roach:.2f}, MAE: {mae_roach:.2f}")

print("\nModelo para Peces Whitefish:")
print(f"R²: {r2_whitefish:.2f}, MSE: {mse_whitefish:.2f}, EMSE: {emse_whitefish:.2f}, MAE: {mae_whitefish:.2f}")


print("Modelo para Peces Smelt:")
print(f"R²: {r2_smelt:.2f}, MSE: {mse_smelt:.2f}, EMSE: {emse_smelt:.2f}, MAE: {mae_smelt:.2f}")


print("\nModelo para Peces Parkki:")
print(f"R²: {r2_parkki:.2f}, MSE: {mse_parkki:.2f}, EMSE: {emse_parkki:.2f}, MAE: {mae_parkki:.2f}")


print("\nModelo para todas las especies:")
print(f"R²: {r2_todos:.2f}, MSE: {mse_todos:.2f}, EMSE: {emse_todos:.2f}, MAE: {mae_todos:.2f}")
```
